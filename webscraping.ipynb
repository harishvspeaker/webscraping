```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
final=pd.DataFrame()
for j in range(1,20):
  webpage=requests.get('https://www.flipkart.com/search?q=sneakers+&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  company=soup.find_all('div', class_="_1xHGtK _373qXS")
  brand=[]
  name=[]
  price=[]
  original=[]
  for i in company:
      brand.append(i.find('div' , class_="_2WkVRV").text)
      name.append(i.find('a', class_="IRpwTa").text)
      price.append(i.find('div', class_="_30jeq3").text)
      original.append(i.find('div', class_="_3I9_wc").text)
        
df=pd.DataFrame({'brand':brand,
    'name':name,
    'price':price,
    'original':original,
    
    })
  
final=final.append(df,ignore_index=True)

     
     
    
  
```

```python
final.to_csv('flipkart.csv')
```
